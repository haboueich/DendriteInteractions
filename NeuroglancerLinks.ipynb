```python
# Neuroglancer link for all minimum distances
import cloudvolume
from caveclient import CAVEclient
import nglui

from nglui import statebuilder

client = CAVEclient('minnie65_phase3_v1')
client.info.get_datastack_info()

img, seg = statebuilder.from_client(client)

pt_map = statebuilder.PointMapper('Coordinates', linked_segmentation_column='Seg ID Nrn 2')
anno = statebuilder.AnnotationLayerConfig('soma_pts', mapping_rules=pt_map, linked_segmentation_layer=seg.name,
                                           tags=['', 'BC'])
sb = statebuilder.StateBuilder([img, seg, anno], client=client)

sb.render_state(df3[['Seg ID Nrn 2','Coordinates']], return_as='html')
```

```python
# Neuroglancer link for specific pair
pt_map = statebuilder.PointMapper('Coordinates', linked_segmentation_column='Seg ID Nrn 2')
anno = statebuilder.AnnotationLayerConfig('soma_pts', mapping_rules=pt_map, linked_segmentation_layer=seg.name,
                                         tags=['', 'BC'])
sb = statebuilder.StateBuilder([img, seg, anno], client=client)

# Enter the pair you would like to see
A = int(input("Enter 1st neuron (a number between 0 and 162)"))
B = int(input("Enter 2nd neuron (a number between 0 and 162)"))

df2 = df3.loc[(df3['Nrn 1'] == A) & (df3['Nrn 2'] == B), ['Nrn 1', 'Nrn 2', 'Distance', 'Seg ID Nrn 1', 'Seg ID Nrn 2', 'Coordinates']]
print(df2)

sb.render_state(df2[['Seg ID Nrn 2','Coordinates']], return_as='html')
```

```python
# Neuroglancer link for a specific pair in a specific distance
pt_map = statebuilder.PointMapper('Coordinates', linked_segmentation_column='Seg ID Nrn 2')
anno = statebuilder.AnnotationLayerConfig('soma_pts', mapping_rules=pt_map, linked_segmentation_layer=seg.name,
                                            tags=['', 'BC'])
sb = statebuilder.StateBuilder([img, seg, anno], client=client)
    
# Enter the pair you would like to see
A = int(input("Enter 1st neuron (a number between 0 and 162)")) 
B = int(input("Enter 2nd neuron (a number between 0 and 162)")) 
    
# Enter the distance range you would like to see
C = int(input("Enter the minimum distance (a number between 0 and 5)"))
D = int(input("Enter the maximum distance (a number between 0 and 5)"))
    
df4 = df3.loc[(df3['Nrn 1'] == A) & (df3['Nrn 2'] == B) & (df3['Distance'] >= C) & (df3['Distance'] <= D), ['Nrn 1', 'Nrn 2', 'Distance', 'Seg ID Nrn 1', 'Seg ID Nrn 2', 'Coordinates']]
print(df4)

sb.render_state(df4[['Seg ID Nrn 2', 'Coordinates']], return_as='html')
```
