```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_feather('data_frame3.feather') # DataMatrix1
df1
```

```python
CellTypesWithIDs = pd.read_feather('CellTypesWithIDs.feather')
CellTypesWithIDs
```

```python
# Sort the Dataframe by distance
SortedDistance = df1.sort_values(by=['Distance'])
SortedDistance
```

```python
SortedDistanceAll = pd.merge(SortedDistance, CellTypesWithIDs, left_on = 'Seg ID Nrn 1', right_on = 'pt_root_id', how = 'left')
```

```python
SortedDistanceAll2 = pd.merge(SortedDistanceAll, CellTypesWithIDs, left_on = 'Seg ID Nrn 2', right_on = 'pt_root_id', how = 'left')
DistanceAll = SortedDistanceAll2[["Nrn 1", "Nrn 2", "Distance", "Seg ID Nrn 1", "Seg ID Nrn 2", "predicted_MET-type Label_x", "predicted_MET-type Label_y"]]    
df2 = pd.DataFrame(DistanceAll, columns=['Nrn 1', 'Nrn 2', 'Distance', 'Seg ID Nrn 1', 'Seg ID Nrn 2', 'predicted_MET-type Label_x', 'predicted_MET-type Label_y'])
df2.reset_index(drop=True).to_feather("DistanceAll.feather")
df2
```

```python
# Step 1: Enter the minimum and maximum distance you would like to plot
A = int(input("Enter the minimum distance (a number between 0 and 5)"))
B = int(input("Enter the maximum distance (a number between 0 and 5)"))

# Step 2: Extract the number of neuron-to-neuron interactions within the range entered
df3 = df2.loc[(df2['Distance'] >= A) & (df2['Distance'] <= B), ['Nrn 1', 'Nrn 2', 'Distance', 'Seg ID Nrn 1', 'Seg ID Nrn 2', 'predicted_MET-type Label_x', 'predicted_MET-type Label_y']]
df3
```

```python
# Step 3: Convert to an array & inverse the cell types of Nrn 1 and Nrn 2 before sorting
# Convert to an array
DistanceAllArray = df3.to_numpy()

# Array with 7 columns: Nrn 1, Nrn 2, Distance, Seg ID Nrn 1, Seg ID Nrn 2, Cell Type 1, Cell Type 2
# r, c = 7, len(df3) # *2 
# Distances = [[0 for x in range(r)] for y in range(c)]

print(len(DistanceAllArray), "******")
i = 0
while i < len(DistanceAllArray):
    if str(DistanceAllArray[i][5]) > str(DistanceAllArray[i][6]): # if cell type 1 > cell type 2 i.e. alphabetically higher, then switch them
        temp = DistanceAllArray[i][0]
        DistanceAllArray[i][0] = DistanceAllArray[i][1]
        DistanceAllArray[i][1] = temp
        temp = DistanceAllArray[i][5]
        DistanceAllArray[i][5] = DistanceAllArray[i][6]
        DistanceAllArray[i][6] = temp
    i += 1

# Step 4: Convert to a dataframe & sort by cell type
df4 = pd.DataFrame(DistanceAllArray, columns=['Nrn 1', 'Nrn 2', 'Distance', 'Seg ID Nrn 1', 'Seg ID Nrn 2', 'predicted_MET-type Label_x', 'predicted_MET-type Label_y'])

# Sort the Dataframe by cell type of Nrn 1 and cell type of Nrn 2
SortedbyCellType = df4.sort_values(by=['predicted_MET-type Label_x', 'predicted_MET-type Label_y'])
# Convert to an array
SortedCellTypeArray = SortedbyCellType.to_numpy()
# print(SortedCellTypeArray)

# Create an array that will hold the cell data types and their number of interactions; initialize all entries to 0
# Array with 5 columns: cell type 1, cell type 2, num_interactions, Seg ID Nrn 1, Seg ID Nrn 2
# CELLTYPEPAIRNUM changes for each distance range
# 0-1: 184, 1-2: 193, 2-3: 196, 3-4: 199, 4-5: 200
CELLTYPEPAIRNUM = 200 # This is the number of unique cell pair types
r1, c1 = 5, CELLTYPEPAIRNUM 
CellTypeInteractions = [[0 for x in range(r1)] for y in range(c1)]

# Go through this sorted dataframe one by one to count the number of interactions between each pair of cell types
i = 0
k = 0 
while i < len(SortedCellTypeArray):
    if ((len(str(SortedCellTypeArray[i][5])) == 3) | (len(str(SortedCellTypeArray[i][6])) == 3)): # To ensure the cell types are not empty (nan)
        i += 1
        continue
    if i == 0:
        # print("****")
        CellTypeInteractions[k][0] = SortedCellTypeArray[i][5]   # set to Neuron 1 cell type
        CellTypeInteractions[k][1] = SortedCellTypeArray[i][6]   # set to Neuron 2 cell type
        CellTypeInteractions[k][2] += 1   # increment number of interactions
        CellTypeInteractions[k][3] = SortedCellTypeArray[i][3]
        CellTypeInteractions[k][4] = SortedCellTypeArray[i][4]
    elif ((CellTypeInteractions[k][0] == SortedCellTypeArray[i][5]) & (CellTypeInteractions[k][1] == SortedCellTypeArray[i][6])):
        # print("----")
        CellTypeInteractions[k][2] += 1   # increment number of interactions        
    else:
        # print("++++")
        # print("k = ", k)
        k += 1
        CellTypeInteractions[k][0] = SortedCellTypeArray[i][5]   # set to Neuron 1 cell type
        CellTypeInteractions[k][1] = SortedCellTypeArray[i][6]   # set to Neuron 2 cell type
        CellTypeInteractions[k][2] += 1   # increment number of interactions
        CellTypeInteractions[k][3] = SortedCellTypeArray[i][3]
        CellTypeInteractions[k][4] = SortedCellTypeArray[i][4]
    i += 1
print("k = ", k)

# Convert array CellTypeInteractionsAll to Dataframe
df5 = pd.DataFrame(CellTypeInteractions, columns=['Cell Type 1', 'Cell Type 2', 'num_interactions', 'Seg ID Nrn 1', 'Seg ID Nrn 2'])
df5
```

```python
# Create a new Dataframe that includes current neuron cell type interactions plus their inverse interactions
# Example: Cell Type 1 with Cell Type 2 AND Cell Type 2 with Cell Type 1
r2, c2 = 5, len(CellTypeInteractions)*2 
CellTypeInteractionsAll = [[0 for x in range(r2)] for y in range(c2)]
i = 0
j = 0
while i < len(CellTypeInteractions):
    CellTypeInteractionsAll[j] = CellTypeInteractions[i]
    j += 1
    CellTypeInteractionsAll[j][0] = CellTypeInteractions[i][1]
    CellTypeInteractionsAll[j][1] = CellTypeInteractions[i][0]
    CellTypeInteractionsAll[j][2] = CellTypeInteractions[i][2]
    CellTypeInteractionsAll[j][3] = CellTypeInteractions[i][4]
    CellTypeInteractionsAll[j][4] = CellTypeInteractions[i][3]
    j += 1
    i += 1  
print(len(CellTypeInteractionsAll))

# Convert array CellTypeInteractionsAll to Dataframe
df6 = pd.DataFrame(CellTypeInteractionsAll, columns=['Cell Type 1', 'Cell Type 2', 'num_interactions', 'Seg ID Nrn 1', 'Seg ID Nrn 2'])
df6
```

```python
# Pivot plot by cell type and number of interactions within specific distance (from user input)
print("This displays the total number of interactions between each cell type within ", A, " and ", B, " microns")
pivot = df6.pivot_table(
    index="Cell Type 1",
    columns="Cell Type 2",
    values="num_interactions",
    fill_value=0,
    )
df = sns.heatmap(pivot)
```
